```python
import cv2
import os
import time


class_name = "stop"


base_dir = r"E:\Macs_Project_2\codes\testfile"

save_dir = os.path.join(base_dir, class_name)
os.makedirs(save_dir, exist_ok=True)

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Camera not opened")
    exit()

count = len(os.listdir(save_dir))
capturing = False
last_saved_time = 0
interval = 0.5   # save one image every 0.5 sec

print("Press 's' to start/stop saving images")
print("Press 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to read camera")
        break

    display = frame.copy()

    cv2.putText(display, f"Class: {class_name}", (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(display, f"Saved: {count}", (20, 80),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    if capturing:
        cv2.putText(display, "CAPTURING...", (20, 120),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        if time.time() - last_saved_time >= interval:
            img_path = os.path.join(save_dir, f"img_{count}.jpg")
            cv2.imwrite(img_path, frame)
            count += 1
            last_saved_time = time.time()

    cv2.imshow("Dataset Capture", display)

    key = cv2.waitKey(1) & 0xFF

    if key == ord('s'):
        capturing = not capturing
        print("Started capturing" if capturing else "Stopped capturing")

    elif key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
